```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

source_csv = '/home/andris/GitHub/data/churn_prediction/user_series.csv'
```

```python
df = pd.read_csv(source_csv)

df.head()
```

```python
churn_array = df.groupby('id')['event_leg'].max()

churn_array.hist()
plt.xlabel('Last month of activity')
plt.ylabel('Number of users')

plt.show()
y = churn_array.value_counts().reset_index().sort_values('index').set_index('index').cumsum()*100/len(churn_array)

plt.plot(range(len(y)),y,'-ok')
plt.xlabel('Last month of activity')
plt.ylabel('% of remaining users')
plt.grid()
plt.show()
```

```python
df_new = df.groupby('id').sum() 
df_new = df_new.drop('event_leg',axis=1)

df_new = df_new[(df_new < df_new.quantile(0.95)).any(axis=1)]

#df_new.hist(figsize=(20,20),bins=100)

for item in df_new.keys():
    
    plt.hist(df_new[df_new[item]>0][item],100)
    plt.title(item)
    plt.ylim([0,1000])
    plt.xlim([0,1000])
    plt.show()
```

```python
# Let's check the discriminatory power between the users that churn and that don't

label = [1 if item == 0 else 0 for item in churn_array.values]

plt.hist(label)
```

```python
df_grp = df.groupby(['id','event_leg']).sum()
df_churn = df.groupby('id')['event_leg'].max()
label = [1 if item == 0 else 0 for item in df_churn.values]
```

```python
df_new = df_grp[df_grp.index.get_level_values(1)==0].join(df_churn)

df_new['last_month'] = [1 if item == 0 else 0 for item in df_new['event_leg']]
df_new['last_month'].hist()
```

```python
# Compare the features between the 2 groups

df_new.filter(regex='q_lvl_0|last_month').boxplot(by='last_month', figsize = (10,10),showfliers=False,showmeans=True)
```

```python
df_new.filter(regex='q_lvl_1|last_month').boxplot(by='last_month', figsize = (10,10),showfliers=False,showmeans=True)
```

```python
df_new.filter(regex='q_lvl_2|last_month|q_lvl_3|q_lvl_4|q_lvl_5|q_lvl_6').boxplot(by='last_month', figsize = (10,10),showfliers=False,showmeans=True)
```

```python
df_new.filter(regex='play_video|view_hint|last_month').boxplot(by='last_month', figsize = (10,10),showfliers=False,showmeans=True)
```

```python
## Interested to find out how many have no activity at all

# Create another column that is the sum of all the columns
df_new['sum_all'] = df_new.filter(regex='q_').sum(axis=1)
df_new['sum_cor'] = df_new.filter(regex='q_lvl_[0-6]_cor').sum(axis=1)
df_new['sum_att'] = df_new.filter(regex='q_lvl_[0-6]').sum(axis=1) - df_new.filter(regex='q_lvl_[0-6]_cor').sum(axis=1) 
```

```python
df_new[['sum_all','sum_cor','sum_att','last_month']].boxplot(by='last_month', showmeans=True, showfliers=False, figsize=(10,10))
```

```python
df_new[df_new['last_month']==0]['sum_all'].hist(bins=100,figsize=(10,10))
df_new[df_new['last_month']==1]['sum_all'].hist(bins=100,alpha=0.5)
plt.xlabel('Total number of attempts')
plt.ylabel('Total number of occurences')
```

## Conclusions

* A large initial drop-off rate after the first month (around 60%)
* View people do questions of level 5 and 6, and view concepts
* A large range in variation in q_lvl_0 and q_lvl_0_correct
* Discriminatory power in the variables in terms of total questions attemted and answered.
