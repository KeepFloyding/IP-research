## Choosing activity proxies

The key variable that is monitored is improvement in exam score between any 2 given years relative to the mean for those years. This can be normalised and defined as follows:

```
exam_improv_year_1-year_2 = (exam_score_year_1)/<exam_score_year_1> - (exam_score_year_2)/<exam_score_year_2>
```

Activity on the IP platform is defined through the aid of a proxy. 

Performance of the school is defined as the ratio of correct questions to the class size. 
Immersion is defined as the ratio of students on the platform relative to class size. 

IP_activity is defined as a weighted sum of the previous 2 indicators relative to the mean. 

```
Performance = # Tot corr/ class size
Immersion = # Tot users/ class size

IP_activity = w1 x Performance/<Performance> + w2 x Immersion/<Immersion>
```

where the weights ***w1*** and ***w2*** are defined as 1.


````python
# ------------------------------------------------------
# ADDING NEW FEATURES AND INDICES
# ------------------------------------------------------

# Computing improvement in exam scores

qu = ['Biology','Mathematics','Chemistry','Physics']

# Defining it as a ratio to the mean 

for item in qu:
    df['exam_improv_2013-2014_' + item] = (df['exam_score|2014_'+item]/df['exam_score|2014_'+item].mean() - df['exam_score|2013_'+item]/df['exam_score|2013_'+item].mean())

for item in qu:
    df['exam_improv_2013-2016_' + item] = (df['exam_score|2016_'+item]/df['exam_score|2016_'+item].mean() - df['exam_score|2013_'+item]/df['exam_score|2013_'+item].mean())
    
# Student activity
df['Performance'] =  (df.filter(regex='_cor_std').sum(axis=1)/df['Entries|2016_Physics']).fillna(0)
df['Immersion'] =  (df['user_id_std']/df['Entries|2016_Physics']).fillna(0) 

# Trying a combination
df['IP_activity'] = (df['Performance']/df['Performance'].mean()) + (df['Immersion']/df['Immersion'].mean())
````

## Sensitivity analysis

The threshold for IP_activity that seperates schools into active and inactive categories is arbitrary. 
Therefore it is of significant importance to ascertain whether a positive relationship between IP_activity and improvement in exam performance is maintained. 

To ascertain the difference between the active and inactive IP schools, different parametric and non-parametric statistical tests are done:

* Mann-Whitney U test
* Kolmogorov-Smirnov test
* ANOVA (multi-variate f test)


````python
# Checking with a Mann Whitney test

bins = [-0.1, 0, 1, 1000]
cut_off_array = np.array(range(1,100))*0.1

df_score = stat.sensitivity_on_bin_non_para(df, 'IP_activity', df.filter(regex='exam_improv').columns, bins, ['Inactive','Semi-active','Active'], 
                                   cut_off_array, test_type=mannwhitneyu)

stat.plot_sensitivity(df_score,np.unique(df_score['Key']),['p values', 'Group_2 - Group_0'])

# Plotting a line for 5% significance
for item in [1,3,5,7,9,11,13,15]:
    plt.subplot(8,2,item)
    plt.plot([0,10],[0.05,0.05])
````

````python
# Checking with a Mann Whitney test

bins = [-0.1, 0, 1, 1000]
cut_off_array = np.array(range(1,100))*0.1

df_score = stat.sensitivity_on_bin_non_para(df, 'IP_activity', df.filter(regex='exam_improv').columns, bins, ['Inactive','Semi-active','Active'], 
                                   cut_off_array, test_type=ks_2samp)

stat.plot_sensitivity(df_score,np.unique(df_score['Key']),['p values', 'Group_2 - Group_0'])

# Plotting a line for 5% significance
for item in [1,3,5,7,9,11,13,15]:
    plt.subplot(8,2,item)
    plt.plot([0,10],[0.05,0.05])
````

````python
# Checking with a Mann Whitney test

bins = [-0.1, 0, 1, 1000]
cut_off_array = np.array(range(1,100))*0.1

df_score = stat.sensitivity_on_bin(df, 'IP_activity', df.filter(regex='exam_improv').columns, bins, ['Inactive','Semi-active','Active'], 
                                   cut_off_array, test_type='ANOVA')

stat.plot_sensitivity(df_score,np.unique(df_score['Key']),['p values', 'Group_2 - Group_0'])

# Plotting a line for 5% significance
for item in [1,3,5,7,9,11,13,15]:
    plt.subplot(8,2,item)
    plt.plot([0,10],[0.05,0.05])
````

## Checking interpretability 

The metric that defines the performance of the school is abtruse and it is not clear how it translates to the benefit of the student.

This section looks at evaluating how this improvement translates into improvement in exam scores. 
To this, we simply take the cut off that maximises the difference in performance between active and inactive schools and then look at the increase in exam_score relative to each group.



````python
# Choose cut off that maximises the difference value

max_df = df_score[df_score['Group_2 - Group_0'] == df_score.max()['Group_2 - Group_0']]
best_cut_off = max_df['Cut off'].values[0]
best_score = max_df['Group_2 - Group_0'].values[0]

print('Best cut off is', best_cut_off,'with a difference of ',best_score)

# Binning according the cut-off

bins = [-0.1, 0, best_cut_off,10000]
group_names = ['Inactive','Semi-active','Active']

df['categories'] = stat.bin_groups(df,'IP_activity',bins,group_names)

# Checking the exam score of active and inactive groups

inact_2016 = df[df['categories']=='Inactive']['exam_score|2016_Physics'].mean()
inact_2014 = df[df['categories']=='Inactive']['exam_score|2014_Physics'].mean()

act_2016 = df[df['categories']=='Active']['exam_score|2016_Physics'].mean()
act_2014 = df[df['categories']=='Inactive']['exam_score|2014_Physics'].mean()

# Printing the results

print('\n')
print('Exam score in 2014 for active:',act_2014)
print('Exam score in 2016 for active:',act_2016)

print('\n')
print('Exam score in 2014 for inactive:',inact_2014)
print('Exam score in 2016 for inactive:',inact_2016)

print('\n')
print('% increase relative to the mean:', (act_2016-act_2014)*100/act_2014-(df['exam_score|2016_Physics'].mean() - df['exam_score|2016_Physics'].mean())*100/df['exam_score|2016_Physics'].mean())

df['categories'].value_counts().plot(kind='bar')
df['categories'].value_counts()
````

# Developing an analogous scenario to understand the results

Hypothetical scenario to understand this improvement in exam score;
If a total group of 5000 students had a C grade, their initial exam score would be 0.5. 
An improvement of 13% would mean that the exam score changes to 0.5*1.13 = 0.565;

## Number of additional Bs from Cs

This would equate to 
e_0  = 3N_tot/6N_tot = 0.5
e_f = 0.5*1.13 = 0.565

(4*NB + 3 *(N_tot - NB))/(6*N_tot) = e_0*1.13

NB = (e_0*1.13 - 0.5)*6*N_tot
NB = (0.13*0.5)*6*N_tot
NB = 0.39*N_tot

## Number of additional As from Bs

This would equate to 
e_0 = 4N_tot/6N_tot = 0.66

(5*NA + 4 *(N_tot - NA))/(6*N_tot) = e_0*1.13

NA = (e_0*1.13-e_0)*6*N_tot
NA = (0.66*0.13)*6*N_tot
NA = 0.52*N_tot
